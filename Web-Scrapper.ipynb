```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import urllib.request
import time
from selenium.webdriver.common.action_chains import ActionChains


# Set Chrome options for MacBook M1
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-extensions")
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--window-size=1280,800")

# Set up Selenium webdriver with the configured options
driver = webdriver.Chrome(options=chrome_options)

# Navigate to the webpage
driver.get('https://www.bnm.gov.my/-/monthly-highlights-statistics-in-march-2023')  # Replace with the actual URL of the webpage containing the files
time.sleep(5)
# Find all matching elements using CSS_SELECTOR
elements = driver.find_elements(By.CSS_SELECTOR, 'td > a[href^="/documents"]')
# Iterate through the elements and download the files
for element in elements:
    file_url = element.get_attribute('href')
    file_name = file_url.split('/')[-1]  # Extract the file name from the URL

    # Click on the link to trigger the download
    element.click()
    time.sleep(2)  # Wait for the download to start (adjust as needed)

    print(f"Downloaded: {file_name}")

# Quit the browser
driver.quit()
```
